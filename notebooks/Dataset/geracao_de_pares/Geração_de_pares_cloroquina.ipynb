# **Pares Tweets genéricos**

```python
generics = df.loc[(df["is_quote"] == False) & (df["is_reply"] == False)]
generics
```

```python
#gerando pares de tweets generico - Por amostra
sample_size = 10000
pair_generics = [] #lista de dicionario correspondente aos pares
list_tweet_generics = generics["text"].tolist() #transformando tweets em lista

for i in range(sample_size):
  pair_gen = random.sample(list_tweet_generics, 2)
  pair_generics.append({"tweet_1": pair_gen[0], "tweet_2": pair_gen[1], "Class": 0, "Conflict": np.nan})

pair_generics_df = pd.DataFrame(pair_generics) #transformando a lista de dicionarios em df
pair_generics_df
```

```python
#exibindo uma pequena amostra

for i in range(10):
  #printando na ordem - Replied tweet - tweet_original
  print(f'Tweet_1: {pair_generics_df.iloc[i]["tweet_1"]}')
  print(f'Tweet_2: {pair_generics_df.iloc[i]["tweet_2"]}')
  print()
```

```python
#convertendo para csv
pair_generics_df.to_csv("Cloroquina_pares_genericos.csv", index=False, encoding="utf-8")
files.download("Cloroquina_pares_genericos.csv")
```

# **Pares Tweet-Reply**

```python
#filtrando apenas replys cujos tweets respondidos foram encontrados no dataset
replys = df.loc[(df["is_reply"] == True) & (df["reply_text"].notna())]
replys
```

```python
#Gerando pares (por reply)
pair_replys = [] #lista de dicionario que irá armazenar os pares
list_tweet_replys = replys["text"].tolist() #transformando tweets do df em lista de tweets
list_replys = replys["reply_text"].tolist() #transformando replys do df em lista de replys

for i in range(len(list_tweet_replys)):
  pair_replys.append({"tweet_1":  list_replys[i], "tweet_2": list_tweet_replys[i], "Class": 1, "Conflict": np.nan})

pair_replys_df = pd.DataFrame(pair_replys)
pair_replys_df
```

```python
for i in range(10):
  #printando na ordem - Replied tweet - tweet_original
  print(f'Tweet_1: {pair_replys_df.iloc[i]["tweet_1"]}')
  print(f'Tweet_2: {pair_replys_df.iloc[i]["tweet_2"]}')
  print()
```

```python
#convertendo para csv
pair_replys_df.to_csv("Cloroquina_pares_reply.csv", index=False, encoding="utf-8")
files.download("Cloroquina_pares_reply.csv")
```

# **Pares Quote-Tweet**

```python
#filtrando apenas quotes cujos tweets respondidos foram encontrados no dataset
quoteds = df.loc[(df["is_quote"] == True) & (df["quoted_text"].notna())]
quoteds
```

```python
#Gerando pares (por quote)
pair_quote = [] #lista de dicionario que irá armazenar os pares
list_tweet_quote = quoteds["text"].tolist() #transformando tweets do df em lista de tweets
list_quote = quoteds["quoted_text"].tolist() #transformando quote do df em lista de quotes

for i in range(len(list_tweet_quote)):
  pair_quote.append({"tweet_1": list_quote[i], "tweet_2": list_tweet_quote[i] , "Class": 2, "Conflict": np.nan})

pair_quote_df = pd.DataFrame(pair_quote)
pair_quote_df
```

```python
for i in range(10):
  #printando na ordem - Quoted tweet - tweet
  print(f'Tweet_1: {pair_quote_df.iloc[i]["tweet_1"]}')
  print(f'Tweet_2: {pair_quote_df.iloc[i]["tweet_2"]}')
  print()
```

```python
#convertendo para csv
pair_quote_df.to_csv("Cloroquina_pares_quote.csv", index=False, encoding="utf-8")
files.download("Cloroquina_pares_quote.csv")
```

```python
#concatenando os datasets parciais:
df_cloroquina = pd.concat([pair_generics_df, pair_replys_df, pair_quote_df], ignore_index=True, axis = 0)
df_cloroquina
```

```python
#convertendo para csv
df_cloroquina.to_csv("Cloroquina_pares.csv", index=False, encoding="utf-8")
files.download("Cloroquina_pares.csv")
```
